```python
""" DESANIDADO COLUMNA GENRES"""

# Función para convertir cadenas a listas de diccionarios, ignorando valores no válidos
def safe_eval(val):
    if pd.isna(val):
        return []
    if isinstance(val, str):
        try:
            return ast.literal_eval(val)
        except (ValueError, SyntaxError):
            return []
    return []

# Aplicar la función safe_eval a la columna 'genres'
Data_peliculas_copy['genres'] = Data_peliculas_copy['genres'].apply(safe_eval)
# Expandir la columna 'genres' en filas
genres_total = Data_peliculas_copy.explode('genres')

# Filtrar filas donde 'genres' es un diccionario válido
genres_total = genres_total[genres_total['genres'].apply(lambda x: isinstance(x, dict))]

# Separar las claves 'id' y 'name' de los diccionarios en columnas separadas
genres_total['genres_id'] = genres_total['genres'].apply(lambda x: x['id'])
genres_total['genres_name'] = genres_total['genres'].apply(lambda x: x['name'])

# Eliminar la columna original 'genres'
genres_total = genres_total.drop(columns=['genres'])

# Agrupar por película y concatenar los géneros
grouped_genres = genres_total.groupby('id').agg({
    'genres_id': lambda x: ', '.join(map(str, x)),
    'genres_name': lambda x: ', '.join(x)
}).reset_index()

# Renombrar las columnas para mayor claridad
grouped_genres.columns = ['movie_id', 'genres_ids', 'genres_names']

# Ver el resultado para verificar
print(grouped_genres.head())

# fusionar estos datos de nuevo con el DataFrame original
Data_peliculas_copy = pd.merge(Data_peliculas_copy, grouped_genres, left_on='id', right_on='movie_id', how='left')

# Eliminar la columna original 'genres' y la columna de ID de película usada para la fusión
Data_peliculas_copy = Data_peliculas_copy.drop(columns=['genres', 'movie_id'])

# Verificar columnas en Data_peliculas_copy
print("Columnas en Data_peliculas_copy antes de eliminar:")
print(Data_peliculas.columns)

# Verificar si 'movie_id' y 'genres' existen en el DataFrame
columns_to_drop = ['genres', 'movie_id']
existing_columns = [col for col in columns_to_drop if col in Data_peliculas_copy.columns]



print("Columnas en Data_peliculas_copy después de eliminar:")
print(Data_peliculas_copy.columns)
```

```python
""" DESANIDAR COLUMNAS SPOKEN"""

# Aplicar la función safe_eval a la columna 'spoken'
Data_peliculas_copy['spoken_languages'] = Data_peliculas_copy['spoken_languages'].apply(safe_eval)
# Expandir la columna 'spoken' en filas
spoken_total = Data_peliculas_copy.explode('spoken_languages')

# Filtrar filas donde 'spoken' es un diccionario válido
spoken_total = spoken_total[spoken_total['spoken_languages'].apply(lambda x: isinstance(x, dict))]

# Separar las claves 'id' y 'name' de los diccionarios en columnas separadas
spoken_total['spoken_ab'] = spoken_total['spoken_languages'].apply(lambda x: x['iso_639_1'])
spoken_total['spoken_languages_movie'] = spoken_total['spoken_languages'].apply(lambda x: x['name'])

# Eliminar la columna original 'spoken'
spoken_total = spoken_total.drop(columns=['spoken_languages'])
# Agrupar por película y concatenar los spoken
grouped_spoken = spoken_total.groupby('id').agg({
    'spoken_ab': lambda x: ', '.join(map(str, x)),
    'spoken_languages_movie': lambda x: ', '.join(x)
}).reset_index()

# Renombrar las columnas para mayor claridad
grouped_spoken.columns = ['movie_id', 'cod_langauges', 'languages_movie']

# Ver el resultado para verificar
print(grouped_spoken.head())

# fusionar estos datos de nuevo con el DataFrame original
Data_peliculas_copy = pd.merge(Data_peliculas_copy, grouped_spoken, left_on='id', right_on='movie_id', how='left')

# Eliminar la columna original 'genres' y la columna de ID de película usada para la fusión
Data_peliculas_copy = Data_peliculas_copy.drop(columns=['spoken_languages', 'movie_id'])
```

```python
""" DESANIDAR BELONGS TO COLLECTION"""


# Función para evaluar el contenido de la columna
def safe_eval(val):
    try:
        return ast.literal_eval(val) if isinstance(val, str) else val
    except:
        return val


# Aplicar la función safe_eval a la columna 'belongs_to_collection'
Data_peliculas_copy['belongs_to_collection'] = Data_peliculas_copy['belongs_to_collection'].apply(safe_eval)

# Verificar tipos de datos después de safe_eval
print("Tipos después de safe_eval:")
print(Data_peliculas_copy['belongs_to_collection'].apply(type).unique())

# Filtrar filas donde 'belongs_to_collection' es un diccionario válido
df_valid_dicts = Data_peliculas_copy[Data_peliculas_copy['belongs_to_collection'].apply(lambda x: isinstance(x, dict))]

# Expandir la columna 'belongs_to_collection' en filas (esto puede no ser necesario si no es una lista de diccionarios)
collection_total = df_valid_dicts.copy()  # No necesitamos explode si no es una lista de diccionarios

# Separar las claves 'id', 'name', 'poster_path', y 'backdrop_path' en columnas separadas
collection_total['collect_id'] = collection_total['belongs_to_collection'].apply(lambda x: x.get('id', '') if isinstance(x, dict) else '')
collection_total['collect_name'] = collection_total['belongs_to_collection'].apply(lambda x: x.get('name', '') if isinstance(x, dict) else '')
collection_total['collect_poster'] = collection_total['belongs_to_collection'].apply(lambda x: x.get('poster_path', '') if isinstance(x, dict) else '')
collection_total['collect_back'] = collection_total['belongs_to_collection'].apply(lambda x: x.get('backdrop_path', '') if isinstance(x, dict) else '')

# Eliminar la columna original 'belongs_to_collection'
collection_total = collection_total.drop(columns=['belongs_to_collection'])

# Fusionar estos datos de nuevo con el DataFrame original
Data_peliculas_copy = pd.merge(Data_peliculas_copy, collection_total[['id', 'collect_id', 'collect_name', 'collect_poster', 'collect_back']], 
              left_on='id', right_on='id', how='left')

# Verificar los datos procesados
print("Datos extraídos:")
print(collection_total[['collect_id', 'collect_name', 'collect_poster', 'collect_back']].head(10))
```

```python

# Función safe_eval para evaluar cadenas que representan estructuras de datos
def safe_eval(s):
    try:
        return ast.literal_eval(s)
    except (ValueError, SyntaxError):
        return s

# Copia del DataFrame
credits_movies_copy = credits_movies.copy()

# Aplicar la función safe_eval a las columnas 'cast' y 'crew'
credits_movies_copy['cast'] = credits_movies_copy['cast'].apply(safe_eval)
credits_movies_copy['crew'] = credits_movies_copy['crew'].apply(safe_eval)

# Expandir las columnas 'cast' y 'crew' en filas
cast_total = credits_movies_copy.explode('cast')
crew_total = credits_movies_copy.explode('crew')

# Filtrar filas donde 'cast' y 'crew' son diccionarios válidos
cast_total = cast_total[cast_total['cast'].apply(lambda x: isinstance(x, dict))]
crew_total = crew_total[crew_total['crew'].apply(lambda x: isinstance(x, dict))]

# Filtrar solo los directores en la columna 'crew'
crew_total = crew_total[crew_total['crew'].apply(lambda x: x.get('job') == 'Director')]

# Separar las claves 'id' y 'name' de los diccionarios en columnas separadas
cast_total['cast_id'] = cast_total['cast'].apply(lambda x: x['id'])
cast_total['cast_name'] = cast_total['cast'].apply(lambda x: x['name'])

crew_total['crew_id'] = crew_total['crew'].apply(lambda x: x['id'])
crew_total['crew_name'] = crew_total['crew'].apply(lambda x: x['name'])

# Eliminar las columnas originales 'cast' y 'crew'
cast_total = cast_total.drop(columns=['cast'])
crew_total = crew_total.drop(columns=['crew'])

# Agrupar por película y concatenar los valores desanidados
grouped_cast = cast_total.groupby('id').agg({
    'cast_id': lambda x: ', '.join(map(str, x)),
    'cast_name': lambda x: ', '.join(map(str, x))
}).reset_index()

grouped_crew = crew_total.groupby('id').agg({
    'crew_id': lambda x: ', '.join(map(str, x)),
    'crew_name': lambda x: ', '.join(map(str, x))
}).reset_index()

# Renombrar las columnas para mayor claridad
grouped_cast.columns = ['id', 'cast_ids', 'cast_names']
grouped_crew.columns = ['id', 'crew_ids', 'crew_names']

# Unir las transformaciones de cast y crew al DataFrame original
credits_movies_copy = pd.merge(credits_movies_copy, grouped_cast, on='id', how='left')
credits_movies_copy = pd.merge(credits_movies_copy, grouped_crew, on='id', how='left')

# Eliminar las columnas originales 'cast' y 'crew' del DataFrame
credits_movies_copy = credits_movies_copy.drop(columns=['cast', 'crew'])

# Ver el resultado final para verificar
print(credits_movies_copy.head())
```
